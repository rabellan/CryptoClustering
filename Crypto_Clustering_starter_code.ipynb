```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
df_market_data.columns #for scaling the data later
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
df_market_data_array= StandardScaler().fit_transform(df_market_data[['price_change_percentage_24h', 'price_change_percentage_7d',
       'price_change_percentage_14d', 'price_change_percentage_30d',
       'price_change_percentage_60d', 'price_change_percentage_200d',
       'price_change_percentage_1y']])

# Display sample data array for the scaled 7 columns of the df_market_data DataFrame
df_market_data_array
```

```python
# Create a DataFrame with the scaled data
df_market_scaled = pd. DataFrame(df_market_data_array, columns=['price_change_percentage_24h', 'price_change_percentage_7d',
       'price_change_percentage_14d', 'price_change_percentage_30d',
       'price_change_percentage_60d', 'price_change_percentage_200d',
       'price_change_percentage_1y'])

# Copy the crypto names from the original data
# Set the coinid column as index. Call it "Crypto_names" in the new dataframe
# Display sample data
# This df will be used to compute the inertia for every possible value of k later in this notebook
df_market_scaled['Crypto_names'] = df_market_data.index
df_market_scaled = df_market_scaled.set_index('Crypto_names')
df_market_scaled
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_list=list(range(1,11))
k_list
```

```python
# Create an empty list to store the inertia values
inertia =[]

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for k in k_list:
    k_model = KMeans(n_clusters=k, random_state=0)
    k_model.fit(df_market_scaled)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_list, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot = df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve",
    label="Elbow for Original Data",
    xticks=k
)

elbow_plot
# Looks like 4 is a good elbow to start with
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Based on the line graph above, **<span style="color: red;">4</span>** seems to be the point where the inertia starts to get shallow

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the scaled data
# use df_market_scaled
model.fit(df_market_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
kmeans_predictions_array = model.predict(df_market_scaled)

# Print the resulting array of cluster values.
kmeans_predictions_array
```

```python
# Create a copy of the DataFrame
df_market_scaled_with_prediction = df_market_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_scaled_with_prediction['prediction'] = kmeans_predictions_array

# Display sample data
df_market_scaled_with_prediction
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
df_market_scaled_with_prediction_plot = df_market_scaled_with_prediction.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="prediction",
    label="Cluster Original Data",
    hover_cols="[Crypto_names]"
)

df_market_scaled_with_prediction_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
# Use df_market_scaled
pca_transformed_array = pca.fit_transform(df_market_scaled)

# View the first five rows of the DataFrame. 
pca_transformed_array[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
# so the sum of the variance ratio is about 89.5%. Is that good enough? Maybe.
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** **<span style="color: red;">89.5%</span>** 

```python
# Create a new DataFrame with the PCA data.
df_pca_transformed = pd.DataFrame(pca_transformed_array, columns=["PCA1", "PCA2", "PCA3"])
#df_pca_transformed

# Copy the crypto names from the original data
df_pca_transformed["Crypto_names"] = df_market_data.index

# Set the coinid column as index
df_pca_transformed=df_pca_transformed.set_index('Crypto_names')

# Display sample data
df_pca_transformed
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_list_for_pca = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia_for_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k_list_for_pca:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(df_pca_transformed)
    inertia_for_pca.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve for PCA
elbow_for_pca = {"k_list_for_pca":k_list_for_pca, "inertia_for_pca": inertia_for_pca}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_for_pca = pd.DataFrame(elbow_for_pca)

df_elbow_for_pca
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_for_pca = df_elbow_for_pca.hvplot.line(
    x="k_list_for_pca",
    y="inertia_for_pca",
    title="Elbow Curve For PCA",
    label="Elbow for PCA",
    xticks=k_list_for_pca
)

elbow_for_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** Based on the line graph above, it looks like **<span style="color: red;">4</span>**  is a good candidate for the best value of "k".


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** Despite using PCA, the best value for **k** is still **<span style="color: red;">4</span>**

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
cluster_model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
cluster_model.fit(df_pca_transformed)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
k_cluster_predict_array = cluster_model.predict(df_pca_transformed)

# Print the resulting array of cluster values.
k_cluster_predict_array
```

```python
# Create a copy of the DataFrame with the PCA data
df_cluster_pca_prediction = df_pca_transformed.copy()

# Add a new column to the DataFrame with the predicted clusters
df_cluster_pca_prediction["pca_prediction"] = k_cluster_predict_array

# Display sample data
df_cluster_pca_prediction
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
cluster_pca_plot=df_cluster_pca_prediction.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="pca_prediction",
    label="Cluster PCA",
    hover_cols=['Crypto_names']
)
cluster_pca_plot
```
